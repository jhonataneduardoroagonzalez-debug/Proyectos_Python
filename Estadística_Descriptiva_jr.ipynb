## **Distribución de Frecuencias**

A continuación, se presenta la construcción de una distribución de frecuencias para cada una de las  variaibles que componen el dataframe, inicialmente.

En primera instancia se presenta la frecuencia $\displaystyle{n_{i}}$ la cual corresponde al conteo o número de veces que aparece el valor o el atributo de la variable

### **Número de integrantes del hogar**

```python
# Distribución de frecuencias
(df
 .groupby(['cavol', 'weight', 'age', 'bph', 'svi'])
 .size()
 .to_frame(name='frequency')
)
```

```python
# Distribución de frecuencias de 'Cavol' agregando las correspondientes frecuencias relativa, porcentual y acumulada

# Calculate descriptive statistics for the original 'cavol' column
print("Máximo:",df['cavol'].max())
print("Mínimo:",df['cavol'].min())
print("Rango:",(df['cavol'].max()-df['cavol'].min()))
print("Número de Intervalos:",math.ceil(1+(3.3*np.log10(len(df['cavol'])))))

# Create intervals and frequency table
df["cavol_group"] = pd.cut(df["cavol"],bins=8) # Using 8 bins as in cell ovwpa4DZCG2F
frequency0 = (df
 .groupby('cavol_group', observed=True)
 .agg(frequency=('cavol_group','count'))
)

# Calculate relative frequency
frequency0['relative_frequency'] = frequency0['frequency'] / frequency0['frequency'].sum()

frequency0['percentage_frequency'] = frequency0['relative_frequency'].round(3)*100
frequency0['cumulative_frequency'] = frequency0['frequency'].cumsum()
frequency0['cumulative_relative_frequency'] = frequency0['relative_frequency'].cumsum().round(3)
frequency0['cumulative_percentage_frequency'] = frequency0['percentage_frequency'].cumsum().round(3)
display(frequency0)
```

```python
modes = [round(float(x), 6) for x in s.mode().tolist()]
res_df = pd.DataFrame({
    "valor": {
        "n": int(s.shape[0]),
        "min": s.min(),
        "max": s.max(),
        "media": s.mean(),
        "mediana": s.median(),
        "moda": ", ".join(map(str, modes)),  # ya redondeada
        "varianza_poblacional": s.var(ddof=0),
        "varianza_muestral": s.var(ddof=1),
        "desv_std_poblacional": s.std(ddof=0),
        "desv_std_muestral": s.std(ddof=1),
        "q1": s.quantile(0.25),
        "q2_mediana": s.quantile(0.50),
        "q3": s.quantile(0.75),
        "rango_intercuartilico": s.quantile(0.75) - s.quantile(0.25),
        "rango": s.max() - s.min(),
        "sesgo": stats.skew(s, bias=False),
        "curtosis_exceso": stats.kurtosis(s, fisher=True, bias=False),
        "curtosis_total": stats.kurtosis(s, fisher=True, bias=False) + 3,
    }
})
# Redondear resultados a 3 decimales
res_df_rounded = res_df.round(3)

# Configurar pandas para mostrar solo 3 decimales
pd.set_option('display.precision', 3)
print(res_df_rounded)
```

```python

# --- Clasificación según el peso de la próstata ---
def clasificar_peso(peso):
    if 0 <= peso < 30:
        return "Normal (0 - 30 gr)"
    elif 30 <= peso < 50:
        return "Levemente aumentada (30 - 50 gr)"
    elif 50 <= peso < 80:
        return "Moderadamente aumentada (50 - 80 gr)"
    elif 80 <= peso < 100:
        return "Muy aumentada / masiva (80 - 100 gr)"
    elif 100 <= peso < 500:
        return "Información a Revisar (> 100 gr)"
    else:
        return "Fuera de rango"

# Crear columna con la clasificación
df["weight classification"] = df["weight"].apply(clasificar_peso)

# Calcular frecuencias
frecuencia_peso = (
    df
    .groupby("weight classification", observed=True)
    .agg(frequency=('weight classification', 'count'))
    .reset_index()
)

# Ordenar las categorías de forma lógica
orden = ["Normal (0 - 30 gr)", "Levemente aumentada (30 - 50 gr)", "Moderadamente aumentada (50 - 80 gr)", "Muy aumentada / masiva (80 - 100 gr)", "Información a Revisar (> 100 gr)"]
frecuencia_peso["weight classification"] = pd.Categorical(frecuencia_peso["weight classification"], categories=orden, ordered=True)
frecuencia_peso = frecuencia_peso.sort_values("weight classification")

# Colores personalizados
colors = ["#4CAF50", "#FFEB3B", "#FF9800", "#F44336", "#000000"]  # verde, amarillo, naranja, rojo, negro

# Crear gráfico de anillo
fig, ax = plt.subplots(figsize=(6,6))
wedges, texts, autotexts = ax.pie(
    frecuencia_peso["frequency"],
    labels=frecuencia_peso["weight classification"],
    colors=colors,
    startangle=90,
    autopct=lambda pct: f"{int(round(pct * frecuencia_peso['frequency'].sum() / 100))}",  # solo conteo
    wedgeprops=dict(width=0.35)
)

# Ajustes estéticos
plt.setp(autotexts, size=10, weight="bold", color="black")
ax.set_title("Clasificación del peso prostático", fontsize=14, fontweight='bold')
ax.set(aspect="equal")
```
