### **NUMERAL 1**

 **Creando modelos lineales simples de la forma**


\begin{equation}
y_i = \beta_0 + \beta_1 x_i + \varepsilon_i
\end{equation}


Recordemos que las estimación para los efectos del modelo son:

\begin{equation}
\hat{\beta}_0=\bar{y}-\hat{\beta}_1\bar{x}
\end{equation}

\begin{equation}
\hat{\beta}_1=\frac{S{xy}}{S_{xx}}
\end{equation}

  Construya modelos lineales simples asumiendo que la variable respuesta corresponde al promedio académico acumulado de los estudiantes y como variables explicativas considere cada una de las variables numéricas como puntaje icfes, horas de trabajo, etc. Posteriormente resuelva:

c. determine el coeficiente de determinación y el coeficiente de derterminación ajustado y concluye como explica la variabilidad el modelo

d. Verifique los supuestos del modelo: normailidad y homocedasticidad

### **Modelo lineal simple para variable edad**
Tomando cuenta la aplicación del modelo lineal, obtenemos:

$\displaystyle{y_i=\beta_0+\beta_1x_i}$

donde:

$y_i: $Promedio académico acumulado de los estudiantes\
$x_i :$Edad de los estudiantes\
$B_0: $Parámetro de modelo intercepto\
$B_i: $Efecto que tiene edad sobre promedio académico de los estudiantes

Para aplicar adecuadamente la fórmula de regresión lineal, obteniendo:

$\displaystyle{y_i=3,66+0,0030*x_i}$

**Conclusión**


*   El promedio académico acumulado que se esperaría de los estudiantes es de 3,66, sin tener en cuenta la edad de los mismos.

* A medida que aumente la edad de los estudiantes, se puede incrementar en un 0,0030 décimas el promedio acumulado de los estudiantes.


 ***Intervalos de confianza a 95%***

 Para determinar los IC al 95% establecemos la siguiente fórmula

$\hat{\beta}_j \mp t_{n-2, \frac{\alpha}{2}} \, SE(\hat{\beta}_j)$

Donde:

* $SE(\hat{\beta}_j): $ Error estándar de estimador $\hat{\beta}_j$.
* $t_{n-2, \frac{\alpha}{2}} \, SE(\hat{\beta}_j):$ Valor crítico de la distibución $t$ con 1648 grados de libertad.

Aplicando la fórmula obtenemos:



$Li$=**3,51 promedio académico**

$Ls$=**3,82 promedio académico**

**Conclusión**

* Se espera que inicialmente el promedio académico, tomando en cuenta la edad esté entre **3,51 y 3,82**, adicionalmente obtenemos que este promedio académico obtenga una afetación entre **0 y 0,009 décimas**

 ***Codeterminación $r^2$y $r^2 ajustado$***\
Los valores de codeterminación $r^2$y $r^2 ajustado$ dan como resultado:

$r^2:$ 0.0004536\
$r^2 ajustado:$  -0.0001529

**Conclusión**
* Tan sólo el 0,04536% (valor de r) del promedio académico de los estudiantes se explica por la edad de los estudiantes, por lo cual no tiene relación lineal entre promedio y edad.
* Al realizar el ajuste, con $r^2$ ajustado  la predicción se desajusta en mayor proporción, lo cual indica que no es posible tener una relación lineal entre promedio y edad.

***Normalidad***

Mediante Sharpiro Wilk obtenemos:

* W: 0,98749
* P-value: $9.465×10^{−11}$

**Conclusión**
* La teoría de normalidad es rechazada debido a que el valor de P-Value es menor a 0,05. Mediante esta información el modelo dado no puede ser confiable para calcular mediante normalidad.

***Homocedasticidad***

Mediante Homocedasticidad obtenemos:

* Bartlett’s K-squared=1.6752
* df =2
* P-value: 0,4328

**Conclusión**
* La teoría de homocedasticidad no es rechazada debido a que el valor de P-Value es mayor a 0,05. Las varianzas de la variable edad son homogéneas y constantes.


```python
%%R
# Determinando las estimaciones de los parámetros beta 0 y beta 1
beta_1<-cov(estudiantes$edad,estudiantes$promedio_academico)/var(estudiantes$edad)
beta_0<-mean(estudiantes$promedio_academico)-beta_1*mean(estudiantes$edad)
cat("Estimación de beta 0:",beta_0,"\n")
cat("Estimación de beta 1:",beta_1,"\n")
```

```python
%%R
# construyendo el modelo lineal
mod_edad<-lm(promedio_academico~edad,data=estudiantes)
summary(mod_edad)
```

```python
%%R
# Modelo 1. Variable respuesta promedio academico y variable explicativa la edad
plot(estudiantes$edad,estudiantes$promedio_academico)
abline(mod_edad,col="red")
```

```python
%%R
# Estimaciones por intervalos de confianza para los parámetros de modelo
# con una confianza del 95%
confint(mod_edad,level=0.95)
```

```python
%%R
# Verificando los supuestos del modelo
# normalidad - test de shapiro wilk
shapiro.test(mod_edad$residuals)
```

```python
%%R
# Verificando el supuesto de homocedastidad o de varianza constante
# Aplicando el test de bartlett para la variables edad y promedio academico
estudiantes$edadG<-cut(estudiantes$edad,breaks=quantile(estudiantes$edad,probs=seq(0,1,0.33)),incluide.lowest=TRUE)
bartlett.test(mod_edad$residuals~estudiantes$edadG)
```

Grafica de R-squared

```python
%%R
# Get the R-squared values for each model
r_squared_values <- c(
  summary(mod_n)$r.squared,
  summary(mod_na)$r.squared,
  summary(mod_nb)$r.squared,
  summary(mod_nc)$r.squared
)

# Assign names to the R-squared values
names(r_squared_values) <- c("mod_n", "mod_na", "mod_nb", "mod_nc")

# Create a bar plot
barplot_result <- barplot(r_squared_values,
                          main = "Comparison of R-squared values for different models",
                          xlab = "Model",
                          ylab = "R-squared",
                          col = "skyblue",
                          ylim = c(min(r_squared_values, 0), max(r_squared_values) + 0.001)) # Adjust ylim for text labels

# Add text labels on top of each bar
text(x = barplot_result, y = r_squared_values, labels = round(r_squared_values, 4), pos = 3, cex = 0.8) # Add text labels
```

Grafica de R_squareda ajustado

```python
%%R
# Adjusted R-squaredvalores para cada modelo
adjusted_r_squared_values <- c(
  summary(mod_n)$adj.r.squared,
  summary(mod_na)$adj.r.squared,
  summary(mod_nb)$adj.r.squared,
  summary(mod_nc)$adj.r.squared
)

#  Adjusted R-squared nombre de los valores
names(adjusted_r_squared_values) <- c("mod_n", "mod_na", "mod_nb", "mod_nc")

# Crear grafico de barras
barplot_result <- barplot(adjusted_r_squared_values,
                          main = "Comparison of Adjusted R-squared values for different models",
                          xlab = "Model",
                          ylab = "Adjusted R-squared",
                          col = "skyblue",
                          ylim = c(min(adjusted_r_squared_values, 0), max(adjusted_r_squared_values) + 0.001)) # Adjust ylim for text labels

text(x = barplot_result, y = adjusted_r_squared_values, labels = round(adjusted_r_squared_values, 4), pos = 3, cex = 0.8) # Add text labels
```
